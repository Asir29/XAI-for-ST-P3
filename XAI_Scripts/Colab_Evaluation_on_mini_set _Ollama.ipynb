```python
import subprocess

def explain_with_ollama(costs_current, costs_planned, model='mistral'):
    prompt = f"""
You are an expert autonomous vehicle assistant.

Current trajectory costs: {', '.join([f'{k}={v:.4f}' for k, v in costs_current.items()])}
Planned trajectory costs: {', '.join([f'{k}={v:.4f}' for k, v in costs_planned.items()])}

Explain in simple terms why the planned trajectory was chosen over the current one.
"""
    result = subprocess.run(
        ["ollama", "run", model],
        input=prompt,
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )
    if result.returncode != 0:
        print("Error:", result.stderr)
        return None
    return result.stdout.strip()

# Example usage
costs_current = {
    "safety": 0.0743,
    "costvolume": 0.0933,
    "comfort": 0.0000,
    "progress": 0.9905
}
costs_planned = {
    "safety": 0.0000,
    "costvolume": 0.0000,
    "comfort": 0.0001,
    "progress": 0.8023
}

explanation = explain_with_ollama(costs_current, costs_planned)
print("LLM Explanation:\n", explanation)
```
